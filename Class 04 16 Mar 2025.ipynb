# Massive Bank Data

```python
data=pd.read_excel('bankdataset.xlsx')
```

```python
data
```

```python
data.info()
```

```python
data_domain=data['Domain'].unique()
```

```python
data_domain
```

```python
data_domain = pd.DataFrame(data_domain)
```

```python
data_domain.to_excel("data_domain.xlsx",index_label="Names")
```

```python
data_date=data['Date'].unique()
```

```python
data_date
```

```python
filtered_date=data[(data["Date"] >= "2022-01-01") & (data["Date"] <= "2022-01-31")]
```

```python
filtered_date
```

```python
filtered_date = pd.DataFrame(filtered_date)
filtered_date.to_excel("filtered_date.xlsx",index_label="Names")
```

```python
filtered_dates=filtered_date.groupby(['Domain','Location'])['Transaction_count'].sum()
```

```python
filtered_dates
```

```python
filtered_dates= pd.DataFrame(filtered_dates)
```

```python
filtered_dates.to_excel("data_domainss.xlsx",index_label="Names")
```
